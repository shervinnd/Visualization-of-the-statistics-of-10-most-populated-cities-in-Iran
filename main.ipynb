```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt


# create a new map with bounding coordinates for Iran
m = Basemap(projection='lcc', resolution='h', 
            lat_0=32, lon_0=53,
            llcrnrlon=43.5, llcrnrlat=25.0,
            urcrnrlon=63.5, urcrnrlat=39.5)

# draw map boundaries and coastline
m.drawcoastlines()
m.drawcountries()
m.drawmapboundary(fill_color='white')
# draw parallels and meridians
m.drawparallels(range(20,50,5),labels=[1,0,0,0])
m.drawmeridians(range(40,70,5),labels=[0,0,0,1])

# add title
plt.title('Populous cities of Iran')

# add scatter points to the map
lons = [ 51.4, 46.26, 52.35, 59.37, 48.49, 48.3, 47.03, 57.1, 49.36, 50.56]
lats = [ 35.7, 38.06, 29.39, 36.19, 31.24, 30.3, 34.23, 30.3, 37.18, 34.49]
m.scatter(lons, lats, latlon=True, marker='o', color='red'  )
# add scatter points to the map with names
names = ['Tehran','Tabriz', 'Shiraz', 'Mashahad', 'Ahvaz', 'Abadan', 'Kermanshah', 'Kerman', 'Rasht', 'Qom' ]
for (lon, lat, name) in zip(lons, lats, names):
    x, y = m(lon, lat)
    plt.scatter(x, y, color='red')
    plt.annotate(name, (x, y), textcoords="offset points", xytext=(5,5), ha='center')


sizes = [8.75, 1.37, 1.24, 2.41, 1.31, 0.8, 0.94, 0.68, 0.7, 1.2]  

# add scatter points to the map with sizes
for (lon, lat, name, size) in zip(lons, lats, names, sizes):
    x, y = m(lon, lat)
    plt.scatter(x, y, s=size, color='red', alpha=0.7)  # set the size of the circle
    plt.annotate(name, (x, y), textcoords="offset points", xytext=(5,5), ha='center')
    
    
plt.show()

fig, ax = plt.subplots(figsize=(14, 8))

counts = sizes
bar_labels = names
bar_colors = ['red', 'red', 'orange', 'orange', 'yellow', 'gold', 'orange', 'yellow', 'orange', 'red']

ax.bar(names, counts, label=bar_labels, color=bar_colors)


ax.set_ylabel('population in million')
ax.set_xlabel('Cities')
ax.set_title('Population of Iran in 2015')
ax.legend(title='Populous cities')

plt.show()
```
